```python
import cv2
import numpy as np
from keras.models import model_from_json

# Load the emotion detection model
json_file_path = "/Users/priyanka/Desktop/emotion/emotiondetector.json"  # Update with the correct path
weights_file_path = "/Users/priyanka/Desktop/emotion/emotiondetector.h5"  # Update with the correct path

# Load model architecture from JSON file
with open(json_file_path, "r") as json_file:
    model_json = json_file.read()
classifier = model_from_json(model_json)

# Load model weights
classifier.load_weights(weights_file_path)
print("Model loaded successfully!")

# Load Haar cascade for face detection
haar_file = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_classifier = cv2.CascadeClassifier(haar_file)

# Labels for emotions
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Start webcam feed
cap = cv2.VideoCapture(0)  # 0 for the default webcam
print("Press 'q' to quit.")

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)

        if np.sum([roi_gray]) != 0:
            roi = roi_gray.astype('float') / 255.0
            roi = np.expand_dims(np.expand_dims(roi, axis=0), axis=-1)  # Match model input shape

            prediction = classifier.predict(roi)[0]
            label = emotion_labels[prediction.argmax()]
            cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Emotion Detector', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
